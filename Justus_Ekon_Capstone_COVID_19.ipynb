**Model** **Development**

```python
#Feature Engineering

China['Confirmed_GrowthRate'] = China['Confirmed'].pct_change() * 100
China['Deaths_GrowthRate'] = China['Deaths'].pct_change() * 100
China['Mortality_Ratio'] = (China['Deaths'] / China['Confirmed']) * 100
China['Recovered_GrowthRate'] = China['Recovered'].pct_change() * 100

# Working with the current populaion of China

China_population = 1419321278
China_total_test = 828508482
China['New_case_per_100k'] = (China['New cases']/ China_population) * 100000
China['Test_per_confirmed_cases'] = China_total_test/ (China['Confirmed'])
China.head()
```

```python
China.info()
```

```python
China_data = China.fillna(0)
China_data.head()
```

```python
China_data['Test_per_confirmed_cases'] = China_data['Test_per_confirmed_cases'].replace([np.inf, -np.inf], 0)

China_data.head()
```

```python
# Select columns with float64 dtype
float64_cols = China_data.select_dtypes(include=['float64']).columns

# Replace inf and -inf with NaN so they can be handled by fillna
China_data[float64_cols] = China_data[float64_cols].replace([np.inf, -np.inf], np.nan)

# Fill NaN values with 0 before converting to int64
China_data[float64_cols] = China_data[float64_cols].fillna(0).astype('int64')

# Verify the changes
China_data.info()
```

```python
China_data['Country/Region'] = 'China'
China_data = China_data.drop('WHO Region', axis=1)
```

```python
China_data.head()
```

```python
#Predictive modeling
x = China_data.drop(columns = ['Date', 'Confirmed'], axis = 1)
y = China_data['Confirmed']
```

```python
#Predictive modeling
x = China_data.drop(columns = ['Date', 'Confirmed', 'Country/Region'], axis = 1) # Drop 'Country/Region'
y = China_data['Confirmed']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
lm = LinearRegression()
lm.fit(x_train, y_train)

predictions = lm.predict(x_test)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
prdictions = model.predict(x_test)
predictions
```

```python
#Model evaluation
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
